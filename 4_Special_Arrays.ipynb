# Special types of Array (filled with specific value)

### 1. Zeros
- It creates an array filled with only zeros.

```python
import numpy as np

ar_zero = np.zeros(3)
ar_zero1 = np.zeros((3,4))

print(ar_zero)
print()
print(ar_zero1)
```

# 2. Ones
- It creates an array filled with Ones.

```python
ar_one = np.ones((2,3))
print(ar_one)
```

# 3. Empty
- It creates an array without initializing values (contains random memory data)

```python
ar_emp = np.empty((3,4))
print(ar_emp)
```

# 4. Identity array (np.eye())
### An identity matrix is a square matrix in which:
- All the elements in the main diagonal are 1.
- All the other elements are 0.

```python
ar_iden = np.eye(3,3)
print(ar_iden)
```

# 5. Range 
- It creates evenly spaced values in a range.
- Syntax:- np.arange(start,stop,step)

```python
ar_ran = np.arange(1,10,2)
print(ar_ran)
```

# 6. Linspace
- It creates evenly spaced numbers between two values.
- Syntax:- np.linspace(start, stop, total_numbers)

```python
ar_lin = np.linspace(0,20,5)
print(ar_lin)
```
