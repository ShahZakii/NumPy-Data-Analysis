# Numpy Creation.
### NumPy is used to work with arrays. The array object in NumPy is called ndarray.
### We can create a NumPy ndarray object by using the array() function.
- First import numpy as np.
- Then make a list and create a second variable and pass that list into the new variable.
- Syntax:- y = np.array(list_name)

```python
import numpy as np
lst = [1,2,3,4,5]
arr = np.array(lst)
print(arr)
```

### List with for loop.

```python
lst = []
for i in range(1,5):
    int_i = int(input(f"Enter {i} number: "))
    lst.append(int_i)
print(np.array(lst))
```
