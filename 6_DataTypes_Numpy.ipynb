# Data Types in Numpy.
 - Numpy has some extra datatypes,and it refer to data types with one character, like i for integer, u for unsigned integer etc.
 

### Checking the Data Type of an array.
- The numpy array object has a property called (dtype) that returns the type of the array.
- Syntax:- print(arr.dtype)

```python
import numpy as np

arr = np.array([1,2,3])
print("Data Type:",arr.dtype)
```

```python
arr_one = np.array([1.4, 6.4, 2.5])
print("Data Type:",arr_one.dtype)
```

```python
arr_two = np.array(["apple", "banana", "mango"])
print("Data Type:",arr_two.dtype)
```

### Creating arrays with defined Data Types.

```python
arr_def = np.array([1,3,5,7], dtype = 'S')
print("Data Type:",arr_def.dtype)
```

```python
arr_def1 = np.array([3,4,5,7])

new_arr = np.float32(arr_def1)
print("Data Type:",new_arr.dtype)
```
