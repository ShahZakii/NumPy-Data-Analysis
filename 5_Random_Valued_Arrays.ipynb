# Generate Random Numbers.
- Numpy offers the random module to work with random numbers.
- Random module helps in to create the arrays filled with random numbers using numpy.random module.

### 1. rand()
- It is used to generate a random value between 0 to 1.
- Syntax:- np.random.rand(size of array)

```python
import numpy as np

var = np.random.rand(4)
var1 = np.random.rand(3,3)

print(var)
print()
print(var1)
```

# 2. randn()
- It is used to generate a random value closer to 0.
- This may returns positive or negative numbers as well.
- Syntax:- np.random.randn(size of array)

```python
var2 = np.random.randn(3,4)
print(var2)
```

# 3. randint()
- It is used to generate a random number between a given range.
- Syntax:- np.random.randint(start, end, total_numbers)

```python
var3 = np.random.randint(5, 50, 10)

print(var3)
```
